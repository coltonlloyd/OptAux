```python
import pandas as pd
import glob
from OptAux import resources, Ale_Resquencing
from collections import defaultdict
```

```python
pair_to_ales = {'hisD_pyrC': [2, 3, 4],
                'hisD_gltA': [9, 10, 11, 12],
                'hisD_gltB': [5, 6, 8]}
```

```python
resource_dir = resources.__path__[0]
df = pd.read_excel('%s/key_mutations_final.xls' % resource_dir).set_index("Gene")
```

```python
def get_genes(df):
    gene_set = set()
    for genes in df.index:
        for gene in genes.split(',  '):
            gene_set.add(gene.strip('[]'))
    return gene_set
```

```python
out = {}
for pair in ['hisD_pyrC', 'hisD_gltA', 'hisD_gltB']:
    out[pair] = {}
    df = pd.read_excel('%s/key_mutations_final.xls' % resource_dir, sheet_name=pair).set_index("Gene")
    for gene in get_genes(df):
        for s in glob.glob('../OptAux/Ale_Resquencing/dups/*'):
            dup_df = pd.read_csv(s, index_col=0)
            if gene in dup_df.index and dup_df.loc[gene, 'Multiplicity']>1.5:
                print(s.split('/')[-1], gene, dup_df.loc[gene, 'Multiplicity'])
                if gene not in out[pair]:
                    out[pair][gene] = defaultdict(int)
                out[pair][gene][s.split('/')[-1].split('_')[2]] += 1
            else:
                pass
```

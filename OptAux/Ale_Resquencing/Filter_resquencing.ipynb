```python
def add_gene_product_to_df(df):
    for genes in df.index:
        function_list = []
        for gene in genes.split(',  '):
            function_list.append(function_dict.get(gene, 'N/A'))
        df.loc[genes, 'Function'] = '/'.join(function_list)
    return df
                                 
```

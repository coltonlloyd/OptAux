```python
import pandas as pd
import glob
import seaborn as sns
%matplotlib inline
```

```python
loc = '/home/sbrg-cjlloyd/OptAux/OptAux/Ale_Resquencing/dups/'
ale_list = []
for fi in sorted(glob.glob(loc + '*.csv')):
    # Filter starting strains
    if '_0_' in fi:
        continue
    if 'B_7' in fi or 'C_1_' in fi or 'pyrC' not in fi:
        continue
    ale = int(fi.split('_')[-3])
    exp = fi.split('/')[-1].replace('.csv', '')
    flask = int(fi.split('_')[-2])
    pair = '_'.join(fi.split('/')[-1].split('_')[:2])
    
    #if flask <= 12:
    #    continue
    print(fi)
    if pair not in ale_list:
        df = pd.DataFrame()
        ale_list.append(pair)
    for gene, mult in pd.read_csv(fi, index_col = 0)['Multiplicity'].iteritems():
        df.loc[gene, exp] = mult
```
